```python
#making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)

#preprocessing the data
data = pd.read_csv('gdData.csv')
X = data.iloc[:,0]
Y = data.iloc[:,1]
plt.scatter(X,Y)
plt.show()
```

```python
#building the model
w=0
b=0
A = 0.0001
iterations =1000

n = float(len(X))

#starting gradient descent process
for i in range(len(X)):
  Y_pred = w*X + b
  D_w = (1/n)*sum(X * (Y_pred - Y))
  D_b = (1/n) * sum(Y_pred - Y)
  #updating w and b simultaneously as required for gradient descent
  w = w- A * D_w
  b = b - A * D_b
print(w, b)
```

```python
# making predictions
Y_pred = w*X + b
plt.scatter(X,Y)
plt.plot(X,Y_pred,color='r')
plt.show()
```
